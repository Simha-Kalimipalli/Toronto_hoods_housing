```python
ccdefault = pd.read_csv("drive/My Drive/development/MachineLearning/ccdefault.csv")




```

```python
 list(ccdefault.columns.values)

ccd = ccdefault[ccdefault.columns[1:]]

list(ccd.columns.values)
```

```python
import numpy as np

test_idx =  np.random.uniform(0,1,len(ccd)) <=0.33
                          
train =ccd[test_idx == True]

test = ccd[test_idx == False]

train.head()


feature = ['LIMIT_BAL', \
 'SEX',\
 'EDUCATION',\
 'MARRIAGE',\
 'AGE',\
 'PAY_0',\
 'PAY_2',\
 'PAY_3',\
 'PAY_4',\
 'PAY_5',\
 'PAY_6',\
 'BILL_AMT1',\
 'BILL_AMT2',\
 'BILL_AMT3',\
 'BILL_AMT4',\
 'BILL_AMT5',\
 'BILL_AMT6',\
 'PAY_AMT1',\
 'PAY_AMT2',\
 'PAY_AMT3',\
 'PAY_AMT4',\
 'PAY_AMT5',\
 'PAY_AMT6',\
 'DEFAULT']
```

```python


n = 5

results = []

clf = KNeighborsClassifier(n_neighbors=n)

clf.fit(train[feature], train['DEFAULT'])
        

preds = clf.predict(test[feature])


accuracy = np.where(preds == test['DEFAULT'], 1, 0 ).sum() / float(len(test))*100

print(accuracy)
```
